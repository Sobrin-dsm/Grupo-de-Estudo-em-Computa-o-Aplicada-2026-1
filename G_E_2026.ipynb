# Entendendo o contexto

Para melhor entender o contexto do banco de dados estudado, foi elaborada uma pesquisa acerca do tema em questão, a qual gerou os seguintes frutos (resumidos):

O contexto é acerca da Citologia Aspirativa (PAAF): Onde um médico inseriu uma agulha fina em uma massa (caroço) na mama, extraiu células e as colocou sob um microscópio. O que há no DataFrame são medições geométricas dessas células.

As 30 colunas são divididas em 10 características básicas, medidas de três formas diferentes:

# 1. As 10 Características Básicas
### Cada uma dessas métricas descreve a aparência física do núcleo das células:

Tamanho: Raio, Perímetro e Área.

Formato: Suavidade, Concavidade, Pontos Côncavos e Simetria.

Complexidade: Textura, Compacidade e Dimensão Fractal (detalhes da borda).

# 2. Como os dados são organizados?
Cada uma das 10 métricas acima aparece em 3 versões:

Média (Mean): A média dessas medidas para todas as células encontradas na amostra do paciente.

Erro (Error): O quanto essas medidas variam entre as células. Se o erro é alto, as células são muito diferentes entre si (o que pode ser um sinal de alerta).

Pior (Worst): A média das 3 maiores medidas encontradas. Na medicina, muitas vezes o "pior caso" é mais importante do que a média geral para detectar um tumor.

# 3. O Objetivo Final
Diferenciar células Benignas (pequenas, lisas e uniformes) de células Malignas (grandes, deformadas, com bordas irregulares e muita variação entre elas).

```python
df['DataAplicacao'] = pd.to_datetime(df['DataAplicacao'], errors='coerce')
```

```python
#Limpeza Geral de resíduos (Remover "1: ", "2: ", etc de todas as colunas de texto)
# Isso limpa Sexo, Raça, Tipo de Evento de uma vez só
cols_para_limpar = df.select_dtypes(include=['object']).columns
for col in cols_para_limpar:
    df[col] = df[col].str.replace(r'^\d+:\s*', '', regex=True).str.strip()
```

```python
# Lista das colunas que precisam ser separadas por dose
colunas_para_separar = [
    'TipoEvento', 'Reação', 'Gravidade',
    'Desfecho', 'Doenças(CID10)', 'Medicamento', 'TipoAtendimento'
]

for col in colunas_para_separar:
    # Separamos por '/' e expandimos para novas colunas (máximo 6 conforme seu padrão)
    df_temp = df[col].str.split('/', expand=True)

    # Renomeamos e adicionamos ao DataFrame original
    for i in range(6):
        nome_nova_coluna = f'{i+1}_{col}'
        # Se a coluna existir no split, adiciona, senão cria como nula
        if i in df_temp.columns:
            df[nome_nova_coluna] = df_temp[i]
        else:
            df[nome_nova_coluna] = None

    # Após separar, podemos apagar a coluna original para limpar o dataset
    df.drop(col, axis=1, inplace=True)
```

```python
# 1. Trocar nulos por 'Não' em UsoDeMedicamentos
df['UsoDeMedicamentos'] = df['UsoDeMedicamentos'].fillna('Não')

# 2. Limpeza de prefixos (1:, 2:) nas novas colunas criadas
# Isso garante que '1: Febre' vire apenas 'Febre'
novas_cols = [c for c in df.columns if any(p in c for p in ['_TipoEvento', '_Reação', '_Gravidade', '_Desfecho', '_Doenças', '_Medicamento', '_TipoAtendimento'])]

for col in novas_cols:
    if df[col].dtype == 'object':
        # Remove o padrão 'Número: Espaço' e espaços em branco
        df[col] = df[col].str.replace(r'^\d+:\s*', '', regex=True).str.strip()
```
